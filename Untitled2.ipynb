```python
import pandas as pd  # pandas をインポート

# エクセルのファイルパス
file_path = "/content/rules2.xlsx"

# エクセルを読み込む
df = pd.read_excel(file_path)

# データを表示
print(df)
```

```python
# ルールをリストに変換
rules = df.to_dict(orient="records")

# ルールを表示（確認用）
for rule in rules:
    print(rule)
```

```python
def check_text(text, rules):
    results = []  # チェック結果を保存するリスト

    for rule in rules:
        if rule["チェック対象"] == "動画・テキスト・音声・画像":  # テキストルールのみ処理
            keyword = rule["NG表現"]
            condition = rule["条件"]

            if condition == "含む":
                if keyword in text:
                    results.append(f"✅ OK: '{keyword}' を含んでいます")
                else:
                    results.append(f"❌ NG: '{keyword}' を含んでいません")

            elif condition == "含まない":
                if keyword in text:
                    results.append(f"❌ NG: '{keyword}' が含まれています")
                else:
                    results.append(f"✅ OK: '{keyword}' は含まれていません")

    return results  # チェック結果を返す
```

```python
# テストするテキスト
text_sample = "たった500円でできる"

# ルールチェックを実行
result = check_text(text_sample, rules)

# 結果を表示
for res in result:
    print(res)
```

```python
import streamlit as st
import pandas as pd

# ルールを読み込む（エクセルを想定）
@st.cache_data
def load_rules():
    df = pd.read_excel("rules2.xlsx")
    return df.to_dict(orient="records")

# ルールに従ってテキストをチェック
def check_text(text, rules):
    results = []
    for rule in rules:
        if rule["チェック対象"] == "テキスト":
            keyword = rule["ルールの内容"]
            condition = rule["条件"]

            if condition == "含む" and keyword in text:
                results.append(f"✅ OK: '{keyword}' を含んでいます")
            elif condition == "含まない" and keyword in text:
                results.append(f"❌ NG: '{keyword}' が含まれています")
    return results

# Streamlit UI部分
st.title("📌 ルールチェックツール")
st.write("🔹 **テキストを入力してください**")

# 入力フォーム
text_input = st.text_area("チェックする文章", "")

# ルールをロード
rules = load_rules()

# ボタンを押したらチェック
if st.button("✅ ルールチェックを実行"):
    if text_input:
        result = check_text(text_input, rules)
        st.subheader("📌 チェック結果")
        for res in result:
            st.write(res)
    else:
        st.warning("⚠️ 文章を入力してください")
```

```python
# ユーザーが手動で入力
text_input = input("チェックする文章を入力してください: ")

# ルールを適用
result = check_text(text_input, rules)

# 結果を表示
print("\n📌 チェック結果:")
for res in result:
    print(res)
```
